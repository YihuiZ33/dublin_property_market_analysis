### Define a function to make the data structured

```python
## Select the columns that I am intersted in, the sale_date, address, county, Postal Code', Price (€)
## Modify the price, delete the "\x80" (the currency symbol) and the comma, change the data type to float
def transform_document(doc):
    
    price_value = doc.get('Price (€)')
    
    if isinstance(price_value, str):
        ## If the value is a string, clean and convert it to float
        price = float(price_value.replace('€', '').replace(',', ''))## clear the euros symbols and the comma
    else:
        ## If the value is already a float, keep it as is
        price = price_value
    
    return {
        'sale_date': doc['Date of Sale (dd/mm/yyyy)'],
        'address': doc['Address'],
        'county': doc['County'],
        'eircode': doc['Postal Code'],
        'price': price
    }
```

```python
structured_data = [transform_document(doc) for doc in houseprice]
```

```python
## Define function return latitude and longitude
def geocode_address(address):
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

## To shorten the running time
def batch_geocode(addresses):
    results = []
    for address in addresses:
        results.append(geocode_address(address))
        ## the number of requests to the API is restricted to a maximum of 10 requests per second per user
        time.sleep(0.1)  
    return results

## Batch size for processing
batch_size = 1000

start_time = time.time() ## check the running time

## Split addresses into batches and process asynchronously
for i in range(0, len(houseprice_df), batch_size):
    batch_addresses = houseprice_df['address_full'].iloc[i:i+batch_size].tolist()
    batch_results = batch_geocode(batch_addresses)
    
    ## Ensure batch_results contains tuples of equal length
    filtered_results = [(None, None) if result[0] is None else result for result in batch_results]
    
    ## Unzip the filtered_results
    latitudes, longitudes = zip(*filtered_results)
    
    ## Assign latitudes and longitudes to DataFrame
    houseprice_df.loc[i:i+len(latitudes)-1, 'latitude'] = latitudes
    houseprice_df.loc[i:i+len(longitudes)-1, 'longitude'] = longitudes

end_time = time.time()
print("Total execution time:", end_time - start_time, "seconds")
```

### Assign garda station

```python
import json
from shapely.geometry import shape, Point
```

```python
# Load GeoJSON file
with open("Garda_Districts_-_An_Garda_Siochana.geojson") as f:
    geodata = json.load(f)
```

```python
geodata['features'][0]['geometry']
```

```python
geodata['features'][0]['properties']['Division']
```

```python
## define the function to assign garda station to each address
def assign_garda_division(df):
    longitude = df['longitude']
    latitude = df['latitude']

    for feature in geodata['features']:
        # Convert GeoJSON geometry to Shapely shape
        polygon = shape(feature['geometry'])
        # Check if property is within the polygon
        if polygon.contains(Point(longitude, latitude)):
            return feature['properties']['Division']  # Name of the assigned Garda division

    return None
```

```python
## Apply the function to assign Garda division to each row in the DataFrame
houseprice_df['garda division'] = houseprice_df.apply(assign_garda_division, axis=1)
```
